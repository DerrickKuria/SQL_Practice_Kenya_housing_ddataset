```
import csv

import pandas as pd
```

```
with open('/content/census1.csv','r') as f:
  census1 = pd.read_csv(f, index_col = 0, encoding = 'utf-8')
```

```
%sql DROP TABLE if exists census1;
%sql PERSIST census1;
%sql SELECT * FROM census1;
```

```


%%sql
select * from census1 limit 10
```

```
# Challenge 1
# Find out the most popular type of household material for urban regions.#

%%sql

SELECT  sum(iron_sheet)Total_ironsheets,sum(tiles)Total_Tiles,sum(concrete)Total_Concrete,sum(asbestos)Total_asbestos,sum(grass)Total_grass,sum(makuti)Total_makuti ,sum(No_Tin)Total_tin,sum(mud_dung)Total_Mud_Dung,sum(others)Total_Others
FROM census1
WHERE urban_rural = 'Urban'
ORDER BY concrete ,iron_sheet,	tiles	,asbestos	,grass,	makuti	,No_Tin	,mud_dung,	others
DESC;
```

```
# Challenge 2
# Find out the least popular type of household material for urban regions.

%%sql
select sum(iron_sheet)Total_ironsheets,sum(tiles)Total_Tiles,sum(concrete)Total_Concrete,sum(asbestos)Total_asbestos,sum(grass)Total_grass,sum(makuti)Total_makuti ,sum(No_Tin)Total_tin,sum(mud_dung)Total_Mud_Dung,sum(others)Total_Others
from census1
where urban_rural = 'Urban';
```

```
# Challenge 5
# Find out the average no. of tiling households

%%sql
SELECT AVG(tiles)Average_tiles 
FROM census1
```

```
# Challenge 6
# Which county has the highest no. of grass households?

#%%sql
#SELECT grass, county
#FROM census1
#ORDER BY grass
#DESC
#LIMIT 1;

#or

%%sql

SELECT max(grass)Grass_houses, county
FROM census1;
```

```
# Challenge 7
# Which top 5 counties have the highest no of households made of iron sheets ?
%%sql
SELECT DISTINCT county , iron_sheet
FROM census1
ORDER BY iron_sheet
DESC
Limit 5;
```

```
# Challenge 8
# What is the average no. of households in Kenya?

%%sql
SELECT ROUND(AVG(household),2)Average_no_of_households
FROM census1
```

```
# Challenge 9
# Which county has the most no. of households made of dung?

%%sql 
SELECT county ,mud_dung No_of_Dung_houses
FROM census1
ORDER BY mud_dung
DESC
LIMIT 1;

#or

#Select max(mud_dung)No_of_Dung_houses, county
#from census1
```

```
# Challenge 10
# In ascending order, which top 3 counties have the highest no. of households with makuti?

%%sql
SELECT county, makuti 
FROM census1
ORDER By makuti
DESC
Limit 3;
```

```
# Challenge 11
# What is the no. of Makuti households in urban regions?

%%sql

SELECT sum(makuti)No_of_Makuti , urban_rural
FROM census1
WHERE urban_rural == 'Urban' 
```

```
# Challenge 12
# What is the no. of Makuti households in the rural regions of Kisumu county?

%%%sql
SELECT sum(makuti)Makuti_Kisumu_Rural
FROM census1
WHERE county ='Kisumu' and urban_rural= 'Rural'
```

```
# Challenge 13
# What percentage of households in Kenya are concrete?

%%sql
SELECT ((sum(concrete)/sum(household)) * 100) percentage_of_concrete
FROM census1
```

```
# Challenge 14
# Which top 3 districts have the highest no. of grass and makuti households?

%%sql
SELECT district,makuti,grass
FROM census1
ORDER BY makuti
DESC
LIMIT 3;
```

```
# Challenge 15 
# What is the households distribution of iron sheets, mud/dung, tin, 
# makuti, grass, sheets, concrete, tiles for Baringa County?

%%sql

SELECT  district, iron_sheet ,tiles	,concrete,	asbestos,grass,makuti,No_Tin,mud_dung,others	
FROM Census1
Where county == 'Baringo';
```
